```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

plt.style.use('seaborn-v0_8-whitegrid')
sns.set_palette('husl')
pd.set_option('display.max_columns', 50)
print('‚úÖ Libraries loaded!')
```

```python
DATA_PATH = Path('../../data/raw/GlobalWeatherRepository.csv')
df = pd.read_csv(DATA_PATH)
print(f'üìä Dataset: {df.shape[0]:,} rows √ó {df.shape[1]} columns')
print(f'üíæ Memory: {df.memory_usage(deep=True).sum() / 1024**2:.1f} MB')
```

## üîç Sample Data

```python
df.head()
```

## üìã Column Types

```python
print('Data Types:')
print(df.dtypes.value_counts())
print(f'\nNumeric: {len(df.select_dtypes(include=[np.number]).columns)}')
print(f'Categorical: {len(df.select_dtypes(include=["object"]).columns)}')
```

## üìà Key Statistics

```python
key_cols = ['temperature_celsius', 'humidity', 'pressure_mb', 'wind_kph', 'precip_mm', 'cloud', 'uv_index']
df[key_cols].describe().round(2)
```

## üåç Country Distribution

```python
fig, ax = plt.subplots(figsize=(10, 6))
df['country'].value_counts().head(15).plot(kind='barh', ax=ax, color=sns.color_palette('husl', 15))
ax.set_xlabel('Number of Records')
ax.set_title('Top 15 Countries by Record Count')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
print(f'Total unique countries: {df["country"].nunique()}')
```

## üå°Ô∏è Temperature Distribution

```python
fig, ax = plt.subplots(figsize=(10, 5))
df['temperature_celsius'].hist(bins=50, ax=ax, color='coral', edgecolor='white')
ax.axvline(df['temperature_celsius'].mean(), color='red', linestyle='--', label=f'Mean: {df["temperature_celsius"].mean():.1f}¬∞C')
ax.set_xlabel('Temperature (¬∞C)')
ax.set_ylabel('Frequency')
ax.set_title('Global Temperature Distribution')
ax.legend()
plt.tight_layout()
plt.show()
```

## üå§Ô∏è Weather Conditions

```python
fig, ax = plt.subplots(figsize=(10, 6))
df['condition_text'].value_counts().head(10).plot(kind='barh', ax=ax, color=sns.color_palette('coolwarm', 10))
ax.set_xlabel('Number of Records')
ax.set_title('Top 10 Weather Conditions')
ax.invert_yaxis()
plt.tight_layout()
plt.show()
print(f'Total unique conditions: {df["condition_text"].nunique()}')
```
